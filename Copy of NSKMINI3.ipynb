```python
# -----------------------------------------
# STUDENT PROGRESS TRACKER SYSTEM
# -----------------------------------------

# PART 6 MENU LOOP (Do-While Simulation)
while True:
    # PART 1: Welcome Screen
    print("-----------------------------------------")
    print("       STUDENT PROGRESS TRACKER SYSTEM     ")
    print("-----------------------------------------")

    name = input("Enter Student Name : ")
    student_class = input("Enter Class : ")
    roll = input("Enter Roll Number : ")

    # PART 2: Enter Subject Marks
    num_sub = int(input("How many subjects do you want to enter? : "))

    marks = []
    total = 0

    for i in range(1, num_sub + 1):
        m = int(input(f"Enter marks for subject {i}: "))
        marks.append(m)
        total += m

    average = total / num_sub

    # PART 3: Grade Calculation
    if average >= 90:
        grade = "A"
    elif average >= 75:
        grade = "B"
    elif average >= 50:
        grade = "C"
    elif average >= 35:
        grade = "D"
    else:
        grade = "Fail"

    # PART 4: Pass/Fail Logic
    result = "PASS"
    for m in marks:
        if m < 35:
            result = "FAIL"
            break

    # PART 5: Display Progress Report
    print("\n-----------------------------------------")
    print("        STUDENT PROGRESS REPORT           ")
    print("-----------------------------------------")
    print(f"Name        : {name}")
    print(f"Class       : {student_class}")
    print(f"Roll No     : {roll}")
    print(f"Total Marks : {total}")
    print(f"Average     : {average:.2f}")
    print(f"Grade       : {grade}")
    print(f"Result      : {result}")
    print("-----------------------------------------\n")

    # PART 6: Menu System
    print("What do you want to do next?")
    print("1. Enter marks again")
    print("2. View report again")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        continue  # restart full system

    elif choice == "2":
        # Only reprint last report (without re-entering marks)
        print("\n-----------------------------------------")
        print("        STUDENT PROGRESS REPORT           ")
        print("-----------------------------------------")
        print(f"Name        : {name}")
        print(f"Class       : {student_class}")
        print(f"Roll No     : {roll}")
        print(f"Total Marks : {total}")
        print(f"Average     : {average:.2f}")
        print(f"Grade       : {grade}")
        print(f"Result      : {result}")
        print("-----------------------------------------\n")

    elif choice == "3":
        print("\nExiting the system... Thank you!")
        break

    else:
        print("Invalid choice! Returning to menu...\n")
```
